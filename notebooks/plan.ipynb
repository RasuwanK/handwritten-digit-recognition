# Creating a model predicting a sequence of digits from an image

# Big Picture

Image → CNN → Sequence Features → BiLSTM → CTC → Digit String

# CNN Model

CNN model extracts a feature map, which is given in the shape **(Batch, Channels, Width, Height)** then it is converted into **(Batch, Width, Features)**

# Reason for using BiLSTM 

To read the digit sequences foward and backward

# CTC Temporal Loss

Doesen't need to have the exact input and output alignment

**Finally digit string is predicted as a sequence**
